`Action`
- Merge games and characters so I can do gender by year.

```python
cleaned_games.merge(cleaned_chars, how = "right", left_on = "game_id", right_on = "game")
games_chars = cleaned_games.merge(cleaned_chars, how = "right", left_on = "game_id", right_on = "game")
games_chars.drop(['game'], axis = 1, inplace = True)
games_chars
```

`Action`
- Find the count of the genders by year, but as a percentage as this can help me visualise better just many of each gender take up a part of the total number of characters.

```python
games_chars.groupby('year')[['gender']].value_counts(normalize = True).mul(100).round(1)
```

`Action`
- Representation of the number of gender by year, not as a percentage.

```python
gender_by_year = games_chars.groupby('year')\
                   .agg(count = ('gender', 'value_counts'))


gender_by_year.sort_values(by='year', ascending = True)
```

`Action`
- Plot a bar chart showing the number of each gender over the years using value_counts() instead of the percentages as when using percentages, the graph looks inflated as there is a higher maximum value with the percentages.

```python
year_gender = games_chars.groupby(['year'])['gender'].value_counts()
year_gender

year_gender.unstack().sort_values(by='year', ascending=True).plot(kind='bar');
```

`Result`
- Overwhemingly, male characters are consistently the largest amount followed by female characters. The year with the smallest difference was 2021. Non binary is often the third highest amount within gender as some games present you a main character without specifying the gender.
- Part two of hypothesis 1 has been proved to be true.
